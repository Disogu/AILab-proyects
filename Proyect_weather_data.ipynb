```python
import glob
wfiles= sorted(glob.glob('/content/drive/MyDrive/Ai_labschool_proyects/Weather_datasets/weather*.txt'))
wfiles
```

```python
weather = pd.concat((pd.read_csv(file,sep='\s+') for file in wfiles), axis= 'columns')
weather
```

```python
weather.drop(columns='X', inplace=True)
weather
```

```python
wmelt= weather.melt(id_vars=['year','month','measure'])
wmelt
```

```python
wmelt.variable= wmelt.variable.replace('X','',regex=True).astype(int)
wmelt
```

```python
wmelt.dropna (inplace=True)
wmelt
```

```python
wmelt.rename(columns={'variable':'day'}, inplace=True)
```

```python
wmelt['date']= pd.to_datetime(wmelt[['year', 'month', 'day']])
wmelt
```

```python
wmelt.drop(['year','month','day'], axis=1 , inplace=True)
```

```python
wmelt
```

```python
wf= wmelt.pivot(index='date', columns='measure', values='value').reset_index()
wf
```

```python
import matplotlib.dates as mdates

for col in ['Max.TemperatureF', 'Min.TemperatureF', 'Mean.TemperatureF']:
    wf[col] = pd.to_numeric(wf[col], errors='coerce')

plt.figure(figsize=(15, 6))

plt.plot(wf['date'], wf['Max.TemperatureF'], label='Max Temperature', marker='o', linestyle='-')
plt.plot(wf['date'], wf['Min.TemperatureF'], label='Min Temperature', marker='o', linestyle='-')
plt.plot(wf['date'], wf['Mean.TemperatureF'], label='Mean Temperature', marker='o', linestyle='-')

plt.gca().xaxis.set_major_locator(mdates.AutoDateLocator())
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.gcf().autofmt_xdate()

plt.xlabel('Date')
plt.ylabel('Temperature (Â°F)')
plt.title('Temperature vs. Date')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
